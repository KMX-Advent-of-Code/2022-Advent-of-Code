## Part 1

```python
def recursive_thing(left, right, i=0):
    print(i)
    i = i + 1
    print(f"comparing {left} and {right}")
    if isinstance(left, int) and isinstance(right, int):
        if left < right:
            return True
        elif left > right:
            return False
        elif left == right:
            return -1
    if isinstance(left, list) and isinstance(right, list):
        print(f"Looping through {left} and {right}")
        try:
            for k in range(len(left)):
                myout = recursive_thing(left[k], right[k], i=i)
                if myout == -1:
                    print(f'continue {k,i}')
                    continue
                elif myout == True:
                    return True
                elif myout == False:
                    return False
            if len(left) < len(right):
                return True
            elif len(left) == len(right):
                pass

        except IndexError:
            return False

    if isinstance(left, int) and isinstance(right, list):
        left = [left]
        return recursive_thing(left, right, i=i)

    elif isinstance(left, list) and isinstance(right, int):
        right = [right]
        return recursive_thing(left, right, i=i)
    return -1
```

## Part 2

```python
def mycomp(x,y):
    out = recursive_thing(x,y)
    if out:
        return -1
    else:
        return 1
```

```python
def flatten_list(list_of_lists):
    return [y for x in list_of_lists for y in x]
```

```python
myflat =flatten_list(myinputs)
```

```python
myflat.append([[2]])
myflat.append([[6]])
```

```python
from functools import cmp_to_key
```

```python
sorted_list = sorted(myflat,key=cmp_to_key(mycomp))
```
