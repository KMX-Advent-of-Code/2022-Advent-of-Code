### Work

Part 1

```python
# For parsing moves
d_them = {'A' : 1, 'B' : 2, 'C' : 3}
d_us = {'X' : 1, 'Y' : 2, 'Z' : 3}

def round_score1(code):
    # What each person played
    them = d_them[code[0]]
    us = d_us[code[2]]
    
    # Evaluation (win/lose/draw)
    # Using that values mod 3 characterize win/loss/draw
    # Get it to -1 = loss, 0 = draw, 1 = win, and then scale appropriately
    ev = (us - them) % 3
    if ev == 2:
        ev = -1
    ev = 3 * (ev + 1)
    
    # Return the eval
    return us + ev
```

```python
sum(apply_func(split(s), round_score1))
```

Part 2

```python
# For parsing moves
# -1/0/1 for lose/draw/win
d_them = {'A' : 1, 'B' : 2, 'C' : 3}
d_us_ev = {'X' : -1, 'Y' : 0, 'Z' : 1}

def round_score2(code):
    them = d_them[code[0]]
    ev_sign = d_us_ev[code[2]]
    ev = 3 * (ev_sign + 1)
    us = (them + ev_sign - 1) % 3 + 1
    
    return us + ev
```

```python
sum(apply_func(split(s), round_score2))
```

### Utilities

```python
# Remove initial/final \n characters
def clean(s):
    return s[1:-1]

# Split at \n characters
# If there are \n\n characters, split into blocks too
def split(s):
    out = [block.split('\n') for block in clean(s).split('\n\n')]
    if len(out) == 1:
        return out[0]
    else:
        return out

# Apply a function to a list or "block" data (2-level list)
def apply_func(data, func):
    if isinstance(data[0], list):
        return [[func(x) for x in block] for block in data]
    else:
        return [func(x) for x in data]

# Split, parsing everything as ints
def split_int(s):
    return apply_func(split(s), int)

# Split, parsing everything as float
def split_float(s):
    return apply_func(split(s), float)
```
