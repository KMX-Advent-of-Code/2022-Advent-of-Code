```python
# part2, still want to use pop because it's fun, so just reverse the popped list again
# Re-initialize stacks to start over.
for move in moves:
    todo = move.split(' ')
    numpop = int(todo[1])
    fromstack = int(todo[3])
    tostack = int(todo[5])
    
    temp = []
    for i in range(numpop):
        temp.append(stacks[fromstack].pop())
    for i in range(numpop):   
        stacks[tostack].append(temp.pop())
        
''.join([x[-1] for x in stacks.values()])         
```
