```python
def config_to_lists(config):
    n = len(config[-1])
    m = len(config)
    
    # For each column (indexed by i in the strings), process the stack
    stacks = []
    for i in range(1, n, 4):
        # Get the stack of boxes from bottom to top
        stack = [config[j][i] for j in range(m-2, -1, -1)]
        
        # Convert append
        stacks += [[box for box in stack if not box == ' ']]
    
    return stacks

def instruction_to_tuple(instruction):
    aux = instruction.split()
    return (int(aux[1]), int(aux[3]) - 1, int(aux[5]) - 1)
```

## Utilities

```python
# Remove initial/final \n characters
def clean(s):
    return s[1:-1]

# Split at \n characters
# If there are \n\n characters, split into blocks too
def split(s):
    out = [block.split('\n') for block in clean(s).split('\n\n')]
    if len(out) == 1:
        return out[0]
    else:
        return out

# Apply a function(s) to a list or "block" data (2-level list)
def apply_func(data, func, nested=False):
    if not isinstance(func, list):
        func = [func]
        
    def _func(x):
        for f in func:
            x = f(x)
        return x
        
    if nested:
        return [[_func(x) for x in block] for block in data]
    else:
        return [_func(x) for x in data]

# Split, parsing everything as ints
def split_int(s):
    return apply_func(split(s), int)

# Split, parsing everything as float
def split_float(s):
    return apply_func(split(s), float)
```
